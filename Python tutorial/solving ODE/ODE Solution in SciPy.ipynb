Here, $df$ is a function that calculates the derivatives of a system of ordinary differential equations (ODEs). The ODE system is given by:

$\frac{dy}{dx}=u$

$\frac{du}{dx}=4*x+10*sin⁡(x)−y$

$z$: A list or array representing the current values of variables (in this case, $y$ and $u$).

$y, u = z$:This line unpacks the values in the $z$ list into the variables $y$ and $u$. For example, if $z$ is $[2, 3]$, then $y$ will be assigned $2$ and $u$ will be assigned $3$.

$dydx = [u, 4*x + 10*sin(x) - y]$:This line calculates the derivatives of $y$ and $u$ with respect to $x$ and stores them in the list $dydx$. In this example, $\frac{dy}{dx}$ is equal to $u$, and  $\frac{du}{dx}$ is equal to $4x+10sin⁡(x)−y$.

$z0 = [0, 2]$:This initializes the initial conditions for the ODE system. $z0$ is a list representing the initial values of $y$ and $u$ at $x = pi$.

$sol$ $=$ $odeint(df, z0, x)$ : This line uses the odeint function from the SciPy library to numerically solve the system of ODEs defined by the $df$ function. It takes the initial conditions $z0$, the ODE function $df$, and the array $x$ representing the values of $x$ at which to solve the ODE. The result is stored in the variable sol.




```python
from numpy import sin,pi,arange
def df(z,x):#Here, df is a function that calculates the derivatives of a system of ordinary differential equations (ODEs). The ODE system is given by:
    y,u=z  #z: A list or array representing the current values of variables (in this case, y and u).
    dydx=[u,4*x+10*sin(x)-y]
    return dydx
z0=[0,2]
x=arange(pi,2*pi,0.1)
print(x)
sol=odeint(df,z0,x)
print(sol)
```

# Damped harmonic motion

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def damp(u,t):
    y,z=u #unpacking u
    f=z
    g=-k*y-lam*z
    return [f,g] #raturns two functions in pack

k=1.0
lam=0.4

uo=[0,1] # initial values:y0=0,zo=1

t=np.linspace(0,20,100)

sol=odeint(damp,uo,t)

Y=sol[:,0] # First column
Z=sol[:,1] # second column

plt.axis([0,20,-0.6,0.8]) # set min,max for x and y axes
plt.plot(t,Y,lw=2)
plt.show()
```

# Forced Vibration

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def forced(u,t):
    y,z=u
    f=z
    g=F*np.cos(w*t)-lam*z-k*y
    return [f,g]

lam=0.5
k=5.0
F=1.0

t=np.arange(0,100,0.1)

u0=[0,1.0]
freq=np.arange(0.5,5,0.02)

A=[]

for w in freq:
    sol=odeint(forced,u0,t)
    Y=sol[:,0]
    amp=np.max(Y[-200:])
    A.append(amp)
    
exact=F/np.sqrt((k-freq**2)**2+(lam**2)*(freq**2))

plt.plot(freq[::2],A[::2],"o",freq,exact,"-")
plt.show()
```
