```python
import numpy as np
import matplotlib.pyplot as plt



f = lambda x, y: y+3
z= lambda x:np.exp(x-2)-3

x = 2
xn = 4
y = -2
h = 0.1
n = int((xn - x) / h)
xp=np.linspace(x,xn,n+1)
yp=np.empty(n+1,float)
yp[0]=y

print("x \t\t y(RK4) \t  K1   \t\t K2 \t\t K3 \t\t K4 \t\t y(analytical)")
print("%f \t %f \t %f \t %f \t %f \t %f \t %f" % (x, y, K1, K2, K3, K4, z(x)))

for i in range(1, n+1):
    K1 = h * f(x, y)
    K2 = h * f(x + (h / 2), y + (K1 / 2))
    K3 = h * f(x + (h / 2), y + (K2 / 2))
    K4 = h * f(x + h, y + K3)
    y += (1 / 6) * (K1 + (2 * K2) + (2 * K3) + K4)
    yp[i]=y
    x += h
    print("%f \t %f \t %f \t %f \t %f \t %f \t %f" % (x, y, K1, K2, K3, K4, z(x)))
    
plt.plot(xp,yp,"ro",xp,z(xp),"g--")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["RK4","Analytical"])
plt.show()
```

# Solving system of odes using RK4 method


$\frac{dx}{dt}$=f(x,y,t)=x-y


$\frac{dy}{dt}$=g(x,y,t)=x+y

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y,t):
    return x-y

def g(x,y,t):
    return x+y

def rk4system(x0,y0,t0,tf,h):
    
    n=int((tf-t0)/h)
    T = np.linspace(t0, tf, n+1)
    
    X=np.zeros(n+1)
    Y=np.zeros(n+1)
    X[0]=x0
    Y[0]=y0
    
    for i in range(n):
        
        k1_x = h * f(X[i], Y[i], T[i])
        k1_y = h * g(X[i], Y[i], T[i])
        
        k2_x = h * f(X[i] + 0.5 * k1_x,Y[i] + 0.5 * k1_y,T[i] + 0.5 * h)
        k2_y = h * g(X[i] + 0.5 * k1_x, Y[i] + 0.5 * k1_y, T[i] + 0.5 * h)
        
        k3_x = h * f(X[i] + 0.5 * k2_x, Y[i] + 0.5 * k2_y, T[i] + 0.5 * h)
        k3_y = h * g(X[i] + 0.5 * k2_x,Y[i] + 0.5 * k2_y, T[i] + 0.5 * h)
        
        k4_x = h * f(X[i] + k3_x, Y[i] + k3_y, T[i] + h)
        k4_y = h * g(X[i] + k3_x, Y[i] + k3_y, T[i] + h)
        
        X[i+1] = X[i] + (k1_x + 2*k2_x + 2*k3_x + k4_x) / 6
        Y[i+1] = Y[i] + (k1_y + 2*k2_y + 2*k3_y + k4_y) / 6
        
    return T,X,Y


# Initial conditions and parameters
x0 = 1
y0 = 0
t0 = 0
tf = 10
h = 0.1


G = rk4system(x0, y0, t0, tf, h)



    
```

```python
print(G[0])
```

```python
# Plot the results
plt.plot(G[0], G[1], label='x(t)')
plt.plot(G[0], G[2], label='y(t)')
plt.xlabel('Time t')
plt.ylabel('Values')
plt.legend()
plt.title('Solution of the system of ODEs using RK4 method')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(z):
    t=z[0]
    x=z[1]
    y=z[2]
    
    dt_dt=1
    dx_dt=x-y
    dy_dt=x+y
    
    return np.array([dt_dt,dx_dt,dy_dt])

def rk4(t0,tf,x0,y0,h):
    
    n=int((tf-t0)/h)
    
    X=np.zeros([3,n+1])
    
    X[:,0]=np.array([t0,x0,y0])
    
    for i in range(1,n+1):
        
        k1 = h * f(X[:, i-1])
        k2 = h * f(X[:, i-1] + 0.5 * k1)
        k3 = h * f(X[:, i-1] + 0.5 * k2)
        k4 = h * f(X[:, i-1] + k3)
        
        
        X[:, i] = X[:, i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        
        
    return X,X[:,i]


    
    
```
