```python
# Function definition for Euler method

def f(x,y):
    return x+y

x0=float(input("Enter value of x0="))
y0=float(input("Enter value of y0="))
h=float(input("Enter value of h="))
n=int(input("Enter value of n=2"))

for i in range(n):
    
    y1=y0+h*f(x0,y0)
    x0=x0+h
    y0=y1
    print("value of y at x",x0,"is=",y0)

```

```python
def f(x,y):
    return -x*(y**2)

x0=float(input("Enter value of x0="))
y0=float(input("Enter value of y0="))
h=float(input("Enter value of h="))
n=int(input("Enter value of n="))

for i in range(n):
    
    yi=y0+h*f(x0,y0)
    x0=x0+h
    y0=yi
    print("value of y at x=",round(x0,2),"is=",round(y0,4))
    
    
```

```python
import matplotlib.pyplot as plt



def f(x,y):
    return x*x

x=float(input("Enter value of x="))
y=float(input("Enter value of y="))
h=float(input("Enter value of h="))
n=int(input("Enter value of n="))

X=[]
Y=[]


for i in range(1,n):
    y+=h*f(x,y)
    x+=h
    X.append(x)
    Y.append(y)
    
    print("The value of i is:",i)
    print("The value of x is",x)
    print("The value of y at",x,"is",y)

    
plt.plot(X,Y)

plt.scatter(X,Y)

plt.show()

    


```

```python
from math import exp

f=lambda x,y:x*y
z =lambda x:exp((x**2)/2)

x=0
xn=2
y=1
h=0.1
n=int((xn-x)/h)

print("x \t\t y(Euler)\t y(Analytical)")
print("%f \t %f \t %f" % (x,y,z(x)))
    

for i in range(n):
    y+=f(x,y)*h
    x+=h
    print("%f \t %f \t %f" % (x,y,z(x)))

    
```
