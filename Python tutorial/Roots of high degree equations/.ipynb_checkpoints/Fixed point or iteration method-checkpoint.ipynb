```python
x=0
for iteration in range(1,101):
    xnew=(2*x**2+3)/5
    #print(iteration, xnew)
    if abs(xnew-x)<0.000001:
        break
    x=xnew
    
print(f"The root: {xnew:.5f}")
print(f"The number of iterations: {iteration:d}")
```

```python
x=1.05
for iteration in range(1,101):
    xnew=(2*x**2+3)/5
    print(iteration, xnew)
    if abs(xnew-x)<0.000001:
        break
    x=xnew
    
print("The root:%0.5f" % xnew)
print("The number of iterations:%d" % iteration)
```
