# Forward difference

```python
f=lambda x: 0.1*(x**5)-0.2*(x**3)+0.1*x -0.2

h=0.008
x=0.1

dff1=(f(x+h)-f(x))/h
dff2= (f(x+2*h)-2*f(x+h)+f(x))/(h**2)

print("f\'(%f)=%f"%(x,dff1))
print("f\'\'(%f)=%f"%(x,dff2))
```

# Central difference  

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 0.1 * (x**5) - 0.2 * (x**3) + 0.1 * x - 0.2

h = 0.008
x = np.linspace(0, 1, 11)

dffc1 = (f(x + h) - f(x - h)) / (2 * h)
dffc2 = (f(x + h) - 2 * f(x) + f(x - h)) / (h**2)

print(dffc1)
print(dffc2)

plt.plot(x, f(x), "-k", label="y")  # Original function
plt.plot(x, dffc1, "--b", label="y'")  # First derivative
plt.plot(x, dffc2, "-.r", label="y''")  # Second derivative

plt.xlabel("x")
plt.ylabel("y, y', y''")
plt.legend()
plt.grid()
plt.show()

```

# Backward difference

```python
f=lambda x: 0.1*(x**5)-0.2*(x**3)+0.1*x -0.2

h=0.05
x=0.1

dffb1=(f(x)-f(x-h))/h
dffb2= (f(x)-2*f(x-h)+f(x-2*h))/(h**2)

print("f\'(%f)=%f"%(x,dffb1))
print("f\'\'(%f)=%f"%(x,dffb2))
```

# Differentiation using scipy

```python
from scipy.misc import derivative

f=lambda x: 0.1*(x**5)-0.2*(x**3)+0.1*x -0.2

print(derivative(f,0.1,0.008,1))
print(derivative(f,0.1,0.008,2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

f = lambda x: 0.1 * (x**5) - 0.2 * (x**3) + 0.1 * x - 0.2

h = 0.008
x = np.linspace(0, 1, 11)

dffc1 = derivative(f,x,h,1)
dffc2 = derivative(f,x,h,2)
print(dffc1)
print(dffc2)

plt.plot(x, f(x), "-k", label="y")  # Original function
plt.plot(x, dffc1, "--b", label="y'")  # First derivative
plt.plot(x, dffc2, "-.r", label="y''")  # Second derivative

plt.xlabel("x")
plt.ylabel("y, y', y''")
plt.legend()
plt.grid()
plt.show()
```
