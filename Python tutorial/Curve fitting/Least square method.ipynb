```python
import numpy as np
from numpy.linalg import solve
import matplotlib.pyplot as plt

x=np.array([3,4,5,6,7,8])
y=np.array([0,7,17,26,35,45])

# Compute the various sums

n=x.size
sx=np.sum(x)
sy=np.sum(y)
sx2=np.sum(x**2)
sxy=np.sum(x*y)

# Construct the matrices

M=np.array([[n,sx],[sx,sx2]])
N=np.array([sy,sxy])

# Solution by linalg module

A,B=solve(M,N)

# PLOTTING

yfit=A+B*x
plt.plot(x,y,"o")
plt.plot(x,yfit)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x=np.array([3,4,5,6,7,8])
y=np.array([0,7,17,26,35,45])

xav=np.mean(x)
yav=np.mean(y)

ssxy=np.sum((x-xav)*(y-yav))
ssxx=np.sum((x-xav)**2)

B=ssxy/ssxx
A=yav-B*xav
yfit=A+B*x

plt.scatter(x,y)
plt.plot(x,yfit)
plt.title("Linear fit",size=18)
plt.xlabel("x",size=16)
plt.ylabel("y",size=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import array,sum,mean

x=array([3,4,5,6,7,8])
y=array([0,7,17,26,35,45])

n=len(x)

a = (mean(y)*sum(x**2)-mean(x)*sum(x*y))/(sum(x**2)-n*mean(x)**2)
b = (sum(x*y)-n*mean(x)*mean(y))/(sum(x**2)-n*mean(x)**2)

print('The straight line equation:')
print('f(x) = (%.4f) + (%.4f)x'%(a,b))


yfit=a+b*x

plt.scatter(x,y)
plt.plot(x,yfit)
plt.title("Linear fit",size=18)
plt.xlabel("x",size=16)
plt.ylabel("y",size=16)
plt.show()

```

```python
x = [3.0, 4.0, 5.0, 6.0, 7.0, 8.0]
y = [0.0, 7.0, 17.0, 26.0, 35.0, 45.0]

n = len(x) # number of data points

sumx = sumx2 = sumxy = sumy = 0


for i in range(n):
    sumx += x[i]
    sumx2 += x[i]**2
    sumxy += x[i]*y[i]
    sumy += y[i]
xmean = sumx / n # mean of x values
ymean = sumy / n # mean of y values

a = (ymean*sumx2 - xmean*sumxy)/(sumx2 - n*xmean**2)
b = (sumxy - n*xmean*ymean)/(sumx2 - n*xmean**2)

print('The straight line equation:')
print('f(x) = (%.3f) + (%.3f)x'%(a,b))


# Convert x to numpy array
import numpy as np
x = np.array(x)

yfit = a + b * x

plt.scatter(x, y)
plt.plot(x, yfit)
plt.title("Linear fit", size=18)
plt.xlabel("x", size=16)
plt.ylabel("y", size=16)
plt.show()
```
