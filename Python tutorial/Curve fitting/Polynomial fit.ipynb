```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([0,1,2,3,4,5])
y=np.array([2,8,14,28,39,62])

n=len(x)

sx=np.sum(x)
sx2=np.sum(x**2)
sx3=np.sum(x**3)
sx4=np.sum(x**4)
sx5=np.sum(x**5)
sx6=np.sum(x**6)
sy=np.sum(y)
sxy=np.sum(x*y)
sx2y=np.sum(x**2*y)
sx3y=np.sum(x**3*y)

S=np.array([[n,sx,sx2,sx3],[sx,sx2,sx3,sx4],[sx2,sx3,sx4,sx5],[sx3,sx4,sx5,sx6]])
T=np.array([sy,sxy,sx2y,sx3y])

Sinv=np.linalg.inv(S)
A,B,C,D=np.dot(Sinv,T)

print('The polynomial equation:')
print('f(x) = (%.6f) + (%.6f)x + (%.6f)x**2 + (%.6f)x**3'%(A,B,C,D))

yfit=A+B*x+C*x**2+D*x**3
plt.plot(x,y,"o")
plt.plot(x,yfit)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2, 8, 14, 28, 39, 62])

# Define the fitting function
def fit_func(x, A, B, C, D):
    return A + B * x + C * x**2 + D * x**3

# Perform the curve fitting
parameters, _ = curve_fit(fit_func, x, y)

# Extract the fitted parameters
A, B, C, D = parameters

# Generate the fitted curve
x_fit = np.linspace(min(x), max(x), 100)
y_fit = fit_func(x_fit, A, B, C, D)

# Plotting
plt.scatter(x, y, label='Data')
plt.plot(x_fit, y_fit, color='red', label='Fitted Curve')
plt.title('Polynomial Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Print the fitted parameters
print("Fitted Parameters:")
print("A:", A)
print("B:", B)
print("C:", C)
print("D:", D)
```

```python
import numpy as np

x=np.arange(6)
y = np.array([2, 8, 14, 28, 39, 62])

m=len(x)
n=3 # Degree of polynomials

A = np.zeros((n+1, n+1))
B = np.zeros(n+1)
a = np.zeros(n+1)

# Loops of system formation

for row in range(n+1):
    for col in range(n+1):
        if row == 0 and col == 0:
            A[row,col] = m
              
            continue
        A[row,col] = np.sum(x**(row+col))
    B[row] = np.sum(x**(row) * y)
    
a = np.linalg.solve(A,B) # solution of the linear system [A]{a}={B}


# Display of the results in polynomial form
print('The polynomial: \n')
print('f(x) =\t %f \t'% a[0])
for i in range(1, n+1):
    print('\t %+f x^%d'%(a[i],i))
```
