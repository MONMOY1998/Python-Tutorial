```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])

plt.plot(x1,y1,mfc="r",marker="o",c="k",label="Non-relativistic")
#plt.legend(loc=9,framealpha=1,edgecolor="black",fontsize=15)
plt.plot(x2,y2,mfc="b",marker="D",c="g",label="Relativistic")

plt.legend(loc=2,framealpha=1,edgecolor="m",fontsize=12)

plt.xlabel("Average Pulse in dyne/cm^2",size=15,color="r",style="italic",weight=700,labelpad=8)
plt.ylabel("Calorie Burnage",size=15,color="orange",style="italic",weight=700,labelpad=8)

plt.title("My Graph",size=20,color="k",weight=700,loc="center")

plt.xlim(-0.5,3.5)
plt.ylim(0,12)


plt.yticks([0,2,4,6,8,10,12],color="k",size=12)
plt.xticks([-0.5,0.0,0.5,1,1.5,2,2.5,3,3.5],color="b",size=12)

plt.tick_params(axis="x", width=2, length=7, color="m", direction="inout", which="both", top=True,labeltop=True)
plt.tick_params(axis="y", width=2, length=7, color="g", direction="inout", which="both", right=True,labelright=True)

#plt.minorticks_on()
plt.grid(axis="both",ls="dotted",lw=2.0,c="c",alpha=1)

plt.show()
```
