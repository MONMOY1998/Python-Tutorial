```python
import pandas as pd
d = pd.read_excel("D:\\class of students in Aliah\\Theory marks\\Book1.xlsx")

df=pd.DataFrame(d)

df["Percentage"]=(df["Total"]/100)*100
df.loc[df["Percentage"]<40,["Grade"]]="Fail"
df.loc[(df["Percentage"]>=40) & (df["Percentage"]<60),["Grade"]]="Average"
df.loc[(df["Percentage"]>=60) & (df["Percentage"]<70),["Grade"]]="Good"
df.loc[(df["Percentage"]>=70) & (df["Percentage"]<80),["Grade"]]="Very Good"
df.loc[(df["Percentage"]>=80) & (df["Percentage"]<90),["Grade"]]="Excellent"
df.loc[(df["Percentage"]>=90) & (df["Percentage"]<100),["Grade"]]="Outstanding"
df
```
